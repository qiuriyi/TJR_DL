```python
input_dx = pd.read_csv('input_dx.csv') ## code count 283
input_px = pd.read_csv('input_px.csv') ## code count 240
input_generid = pd.read_csv('input_generid.csv') ## code count 4
input_revcode = pd.read_csv('input_revcode.csv') ## code count 651
input_stdplac = pd.read_csv('input_stdplac.csv') ## code count 45
input_stdprov = pd.read_csv('input_stdprov.csv') ## code count 131
input_svcscat = pd.read_csv('input_svcscat.csv') ## code count 498
### total code count 1852
input_thercls = pd.read_csv('input_thercls.csv') ## code count 222
input_thergrp = pd.read_csv('input_thergrp.csv') ## code count 30
### total code count 2104
```

```python

Outputy_Raw = pd.read_csv('output_y.csv',header=None)

print(datetime.now())
input_all = pd.concat((input_dx,input_px,input_revcode,input_generid,input_stdplac,input_stdprov,input_svcscat,input_thercls,input_thergrp))
del input_dx
del input_px
del input_generid
del input_revcode
del input_stdplac 
del input_stdprov
del input_svcscat 
del input_thercls
del input_thergrp

fcle = LabelEncoder()
fcle.fit(input_all.values[:,2])
cind = fcle.transform(input_all.values[:,2])
print(datetime.now())
```

```python
print(datetime.now())
inputx_value = np.append(input_all.values,cind[:,None],1)
NofUser = Outputy_Raw.shape[0]

Outputy_value = np.append(np.random.permutation(Outputy_Raw.values),np.arange(NofUser)[:,None],1)
Outputy_value = np.append(Outputy_value,np.round(np.arange(NofUser)/10000)[:,None],1) 

Inputx_df = pd.DataFrame(inputx_value,columns=['eid','days','fcid','fcsc'])
Outputy_df = pd.DataFrame(data=Outputy_value,columns=['eid','label','esc','esct'])

Merged_df = pd.merge(Inputx_df,Outputy_df,on='eid',how='left')
Merged_df = Merged_df.values[np.where(Merged_df['days']<365*2)[0],:]
print(datetime.now())
#np.savez('0702dataprep',Inputx_df=Inputx_df,Outputy_df=Outputy_df,Merged_df=Merged_df)
```

```python
#loadata = np.load('0629dataprep.npz')
Merged_df = np.squeeze(Merged_df)

codecnt = int(np.max(Merged_df[:,3]))+2
snum = int(np.max(Merged_df[:,-2]))+1
Data_Full = np.zeros([snum,codecnt],dtype='uint8')
for i in range(Merged_df.shape[0]):
    Data_Full[Merged_df[i,5],Merged_df[i,3]] = Data_Full[Merged_df[i,5],Merged_df[i,3]]+1
    Data_Full[Merged_df[i,5],-1] = Merged_df[i,4]=='t'
ColNames = [None] * codecnt
for i in range(codecnt-1):
    ColNames[i] = Merged_df[Merged_df[:,3]==i,2][0]
    print(ColNames[i])
ColNames[codecnt-1] = 'label'
#np.savez('0601flatfeature',Data_Full=Data_Full)
DF2D = pd.DataFrame(Data_Full,columns=ColNames)
DF2D.to_csv('0702flatfeature.csv', sep=',',index=False)
```
